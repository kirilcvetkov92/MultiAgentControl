```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
def plot_scores(fn):
    scores = np.load(fn)
    means = []
    print(len(scores))
    for i in range(0,scores.shape[0]):
        means.append((i,np.mean(scores[i-min(i,100):i+1])))
        
    means = np.array(means)
    plt.figure(figsize=(10,5))
    plt.plot(scores,'.',color='gray')
    plt.plot(means[:,0],means[:,1],'g')
    plt.grid()
    plt.xlabel('Episode')
    plt.ylabel('Reward')
    plt.title(fn[:-4])
    plt.legend(['Reward per episode', 'Average reward per 100 episodes'])
    print('Environment solved in', len(means), 'episodes')
```

```python
plot_scores('ddpg_scores_new.npy')
```
